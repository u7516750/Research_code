```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt

# Load FITS file
fits_file = fits.open('capture_0.033_20.033_0.fits')
data = fits_file[0].data


```

```python
# Extract relevant data from the FITS file
image_data = np.array(data)

# Calculate mean and variance for each image
mean_values = (np.median(image_data, axis=0))
variance_values = (np.var(image_data, axis=0))

# Generate photon transfer curve
plt.scatter(mean_values, variance_values)
plt.xlabel('Median Signal')
plt.ylabel('Variance')
plt.ylim(2050, 5000)
#plt.xlimnp.log((1850,2050))
plt.title('Photon Transfer Curve')
plt.show()
```

```python

# Calculate mean and variance for each image
mean_values = (np.median(image_data, axis=0))
variance_values = (np.var(image_data, axis=0))

# Apply data clipping
clip_min_y = (500)  # Set the minimum threshold for clipping
clip_max_y =(30000)  # Set the maximum threshold for clipping
clip_min_x = (2000)  # Set the minimum threshold for clipping
clip_max_x = (14000)  # Set the maximum threshold for clipping



clipped_mean_values = np.clip(mean_values, clip_min_x, clip_max_x)
clipped_variance_values = np.clip(variance_values, clip_min_y, clip_max_y)

# Generate photon transfer curve
plt.scatter(clipped_mean_values, clipped_variance_values)
plt.xlabel('Median Signal')
plt.ylabel('Variance')
plt.title('Photon Transfer Curve')
plt.show()
```
